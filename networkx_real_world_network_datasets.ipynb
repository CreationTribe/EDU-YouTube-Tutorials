```python
import networkx as nx
import matplotlib.pyplot as plt

nx.write_edgelist(nx.gn_graph(11),'data/edgelist.txt')

# gdata = nx.read_edgelist('data/edgelist.txt')
# gdata = nx.read_edgelist("data/edgelist.txt", nodetype=int)
# gdata = nx.read_edgelist("data/edgelist.txt",create_using=nx.DiGraph())
gdata = nx.read_edgelist('data/edgelist.txt')
# gdata = nx.read_edgelist('data/edgelist.txt', data=('weight',float))

# gdata = nx.read_pajek('data/edgelist.net')
# gdata = nx.read_gml('data/edgelist.gml')
# gdata = nx.read_graphml('data/edgelist.graphml')
# gdata = nx.read_gexf('data/edgelist.gexf')

gdata.nodes()

gdata.edges(data=True)

print(nx.info(gdata))

print(nx.is_directed(gdata))
```
